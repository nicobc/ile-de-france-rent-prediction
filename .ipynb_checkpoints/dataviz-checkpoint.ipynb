# Data visualisation
This notebook shows the data visualisations for the property ads I scraped to build my Ile-de-France rent prediction app.

## 1. Imports

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import seaborn as sns
import pandas as pd
import os

import descartes
import geopandas as gpd
from shapely.geometry import Point
```

## 2. Read shapefiles and geolocalised data

### 2.1. Read shapefile for départements in Ile-de-France 

```python
idf_shapefile_path = os.path.join('data', 'shapefiles', 'geoflar-departements.shp')
map_idf = gpd.read_file(idf_shapefile_path)
```

### 2.2. Read shapefile for arrondissements in Paris

```python
paris_shapefile_path = os.path.join('data', 'shapefiles', 'arrondissements.shp')
map_paris = gpd.read_file(paris_shapefile_path)
```

### 2.3. Read geolocalised property ads data for the whole of Ile-de-France

```python
filepath = os.path.join('data', 'locations_2020_10_clean.csv')

df_idf = pd.read_csv(filepath, sep='|')

# Bin rooms & bedrooms
df_idf.loc[df_idf.rooms >= 5, 'rooms'] = '≥5'
df_idf.rooms = df_idf.rooms.astype(str)
df_idf.loc[df_idf.bedrooms >= 3, 'bedrooms'] = '≥3'
df_idf.bedrooms = df_idf.bedrooms.astype(str)

df_idf.head()
```

### 2.4. Subset Ile-de-France dataframe by keeping only data for Paris

```python
df_paris = df_idf.loc[df_idf.dept.astype(str).str.startswith('75'), :].copy()
df_paris.head()
```

## 3. Transform `DataFrame`s into `GeoDataFrame`s

```python
def get_geo_df(df):
    return gpd.GeoDataFrame(df, crs='epsg:4326', geometry=[Point(xy) for xy in zip(df.lon, df.lat)])
geo_df_idf = get_geo_df(df_idf)
geo_df_paris = get_geo_df(df_paris)

# free up some memory by removing the original dataframes
del df_idf, df_paris

# remove arrondissement from dept column in geo_df_idf and use literal département names instead of int
geo_df_idf.dept = geo_df_idf.dept.astype(str).str[:2].astype(int)
dept_dict = {
    75: 'Paris',
    77: 'Seine-et-\nMarne',
    78: 'Yvelines',
    91: 'Essonne',
    92: 'Hauts-de-\nSeine',
    93: 'Seine-Saint-\nDenis',
    94: 'Val-de-Marne',
    95: 'Val-d\'oise'
}
geo_df_idf.dept = geo_df_idf.dept.map(dept_dict)

# replace dept column by arrn in geo_df_paris
geo_df_paris['arrn'] = geo_df_paris.dept.astype(str).str[-2:].str.replace(r'0(\d)', r'\1').apply(
    lambda s: s + 'e arr.' if s != '1' else '1er arr.'
)
geo_df_paris = geo_df_paris.drop('dept', axis=1)
```

## 4. Dataviz

```python
def style_barplot(ax):
    ax.tick_params(left=False)
    for spine in ['top', 'right', 'bottom']:
        ax.spines[spine].set_visible(False)
    ax.spines['left'].set_color((0,0,0,.5))
    ax.set_xticks([])
    ax.set_ylabel('')

def style_boxplot(ax):
    sns.despine(bottom=True, ax=ax)
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.grid(True, axis='x', **{'color': 'darkgray', 'linestyle': ':'})
    plt.xticks(**{'color': 'k', 'alpha': .5})
    plt.tick_params(bottom=False)

def axis_off(ax):
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
def plot_col(column, title, figsize=(16, 10), fmt='{}', pad1=5, pad2=5, pad3=0, kind='bar',
             vmin={'idf': None, 'paris': None}, vmax={'idf': None, 'paris': None}):

    fig = plt.figure(figsize=figsize)

    ax = fig.add_subplot(211)
    ax.set_ylabel("Départements of\nIle-de-France", size=14)
    axis_off(ax)
    ax = fig.add_subplot(212)
    ax.set_ylabel('Arrondissements of Paris', size=14)
    axis_off(ax)
    
    ax = fig.add_subplot(221)
    map_idf.plot(ax=ax, color='lightgray', edgecolor='k')
    cax = fig.add_axes([.17, .5, .265, .02])
    geo_df_idf.plot(ax=ax, column=column, cmap='Spectral_r', alpha=.8, edgecolor=(0,0,0,.2),
                    vmin=vmin['idf'], vmax=vmax['idf'], cax=cax, legend=True,
                    legend_kwds={'orientation': 'horizontal'})
    axis_off(ax)

    ax = fig.add_subplot(222)
    if kind == 'bar':
        data_to_plot = geo_df_idf.groupby('dept')[column].mean().sort_values()
        data_to_plot.plot(kind='barh', ax=ax, alpha=.8, edgecolor=(0,0,0,.2))
        style_barplot(ax)
        for y, x in enumerate(data_to_plot):
            ax.text(x=x+pad1, y=y, s=fmt.format(x), ha='left', va='center')
    elif kind == 'boxplot':
        sns.boxplot(x=column, y='dept', data=geo_df_idf,
                    order=geo_df_idf.groupby('dept')[column].median().sort_values(ascending=False).index,
                    palette='Spectral')
        style_boxplot(ax)
    
    ax = fig.add_subplot(223)
    map_paris.plot(ax=ax, color='lightgray', edgecolor='k')
    cax = fig.add_axes([.17, .1, .265, .02])
    geo_df_paris.plot(ax=ax, column=column, cmap='Spectral_r', alpha=.8, edgecolor=(0,0,0,.2),
                      vmin=vmin['paris'], vmax=vmax['paris'], cax=cax, legend=True,
                      legend_kwds={'orientation': 'horizontal'})
    axis_off(ax)

    ax = fig.add_subplot(224)
    if kind == 'bar':
        data_to_plot = geo_df_paris.groupby('arrn')[column].mean().sort_values()
        data_to_plot.plot(kind='barh', ax=ax, alpha=.8, edgecolor=(0,0,0,.2))
        style_barplot(ax)
        for y, x in enumerate(data_to_plot):
            ax.text(x=x+pad2, y=y, s=fmt.format(x), ha='left', va='center')
    elif kind == 'boxplot':
        sns.boxplot(x=column, y='arrn', data=geo_df_paris,
                    order=geo_df_paris.groupby('arrn')[column].median().sort_values(ascending=False).index,
                    palette='Spectral')
        style_boxplot(ax)

    plt.suptitle(title, size=14, y=.98 + pad3)
    plt.show()
```

```python
fig = plt.figure(figsize=(16, 10))

ax = fig.add_subplot(221)
map_idf.plot(ax=ax, color='lightgray', edgecolor='k')
geo_df_idf.plot(ax=ax, alpha=.5, edgecolor=(0,0,0,.2))
axis_off(ax)
ax.set_ylabel('Départements of\nIle-de-France', size=14, labelpad=50)

ax = fig.add_subplot(222)
geo_df_idf.dept.value_counts().sort_values().plot(kind='barh', ax=ax, alpha=.8, edgecolor=(0,0,0,.2))
style_barplot(ax)
for y, x in enumerate(geo_df_idf.dept.value_counts().sort_values()):
    ax.text(x=x+5, y=y, s=f'{x} ({x/geo_df_idf.shape[0]:.1%})', ha='left', va='center')

ax = fig.add_subplot(223)
map_paris.plot(ax=ax, color='lightgray', edgecolor='k')
geo_df_paris.plot(ax=ax, alpha=.5, edgecolor=(0,0,0,.2))
axis_off(ax)
ax.set_ylabel('Arrondissements of Paris', size=14)

ax = fig.add_subplot(224)
geo_df_paris.arrn.value_counts().sort_values().plot(kind='barh', ax=ax, alpha=.8, edgecolor=(0,0,0,.2))
style_barplot(ax)
for y, x in enumerate(geo_df_paris.arrn.value_counts().sort_values()):
    ax.text(x=x+1, y=y, s=f'{x} ({x/geo_df_paris.shape[0]:.1%})', ha='left', va='center')

plt.suptitle('Distribution of property ads across départements of Ile-de-France and arrondissements of Paris',
             size=14)
plt.show()
```

```python
plot_col('price',
         title='Distribution of monthly rents across property ads in départements of Ile-de-France\nand arrondissements of Paris (in €)',
         figsize=(16, 12), vmin={'idf': None, 'paris': None}, vmax={'idf': 2300, 'paris': 4000}, kind='boxplot')
```

```python
plot_col(
    'surface',
    title='Distribution of property surfaces across property ads in départements of Ile-de-France\nand arrondissements of Paris (in m²)',
    figsize=(16, 12), vmin={'idf': None, 'paris': None}, vmax={'idf': 125, 'paris': 145}, kind='boxplot')
```

```python
fig, ax = plt.subplots(figsize=(20, 6))

dftp = geo_df_idf.groupby('dept')['rooms'].value_counts('normalize').unstack().sort_values('1')
dftp.plot(kind='barh', stacked=True, cmap='Spectral', ax=ax, width=.8)

sns.despine(bottom=True)
plt.tick_params(left=False)
plt.ylabel('')
plt.xticks([])
plt.yticks(**{'fontsize': 14})
plt.legend(frameon=False, loc='center', bbox_to_anchor=(.5, 1.05), fontsize=14, ncol=5)
plt.title('Distribution property ads across département of Ile-de-France and number of rooms',
          size=18, pad=40)

for y in range(len(dftp.index)):
    x = 0
    for i, n in enumerate(dftp.iloc[y]):
        color = 'w' if i in [0, 4] else 'k'
        weight = 'bold' if color == 'w' else 'normal'
        if n > .015:
            plt.text(x=x+(n/2), y=y, s=f'{n:.0%}', ha='center', va='center', size=14, color=color, fontweight=weight)
        x += n
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 6))

dftp = geo_df_idf.groupby('dept')['bedrooms'].value_counts('normalize').unstack().sort_values('0')
dftp.plot(kind='barh', stacked=True, cmap='Spectral', ax=ax, width=.8)

sns.despine(bottom=True)
plt.tick_params(left=False)
plt.ylabel('')
plt.xticks([])
plt.yticks(**{'fontsize': 14})
plt.legend(frameon=False, loc='center', bbox_to_anchor=(.5, 1.05), fontsize=14, ncol=4)
plt.title('Distribution property ads across département of Ile-de-France and number of bedrooms',
          size=18, pad=40)

for y in range(len(dftp.index)):
    x = 0
    for i, n in enumerate(dftp.iloc[y]):
        color = 'w' if i in [0, 3] else 'k'
        weight = 'bold' if color == 'w' else 'normal'
        if n > .015:
            plt.text(x=x+(n/2), y=y, s=f'{n:.0%}', ha='center', va='center', size=14, color=color, fontweight=weight)
        x += n
plt.show()
```
